```python
def sigmod(x):
    return 1/(1 + np.exp(-float(x)))

def sigmod_cal(h):
    z = []*len(h)
    for i in range(len(h)):
        z.append([sigmod(int(h[i]))])
    return np.array(z, dtype=float)
```

```python
def hypothesis(X, theta): 
    return np.dot(X, theta) 

def gradient(x, y, theta): 
    h = hypothesis(x, theta)
    z = sigmod_cal(h)
    grad = x.T.dot(z - y)
    return grad 
  

def loss(X, y, theta, m):
    z = sigmod_cal(hypothesis(X, theta))
    loss_value = -np.sum(y*np.log(z) + (1-y)*np.log(1-z))
    return float(loss_value)
```

```python
def gradientDescent(x, y, learning_rate = 0.001): 
    theta = np.array([[1]]*x.shape[1]) 
    count = 0

    while True:
        l_old = loss(x, y , theta, x.shape[0])
        theta = theta + learning_rate * gradient(x, y, theta)
        l_new = loss(x, y , theta, x.shape[0])
        count += 1
        if count == 100:
            break
    return theta
```
